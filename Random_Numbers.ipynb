## The Logistic Map vs. Pseudo-Randon number generation

```python
import matplotlib.pyplot as plt
```

```python
def function(r,x):
    return r*x*(1-x)
```

```python
r = 4
num = 100000
seed = 0.1
list = [seed]
for i in range(num):
    list.append(function(r,list[-1]))
```

```python
plt.hist(list,bins=10,rwidth=0.85,alpha=.5,density=True)
plt.xlabel('random variable',size=14)
plt.ylabel('probability distribution',size=14)
plt.show()
```

```python
import random as rn
```

```python
# Mersenne Twister algorithm
# cycle size: 2^19937 âˆ’1 ~ 10^6000
```

```python
list_rand = [rn.random() for _ in range(num)]
```

```python
plt.hist(list_rand,rwidth=0.85,alpha=.5,density=True)
plt.xlabel('random variable',size=14)
plt.ylabel('probability distribution',size=14)
plt.show()
```

## Monte Carlo "Hit-or-Miss" Method

```python
import numpy as np
```

```python
def dist(x):
    #return (np.pi/2)*np.sin(x*np.pi)
    return 2*np.sin(x*np.pi)**2
```

```python
x = np.array([.01*i for i in range(101)])
y = np.array([dist(i) for i in x])
```

```python
plt.plot(x,y,linewidth  = 5)
plt.xlabel('random variable',size=14)
plt.ylabel('probability distribution',size=14)
plt.show()
```

```python
num_mc = 1000
#points = np.array([[rn.random(),(np.pi/2)*rn.random()] for _ in range(num_mc)])
points = np.array([[rn.random(),2*rn.random()] for _ in range(num_mc)])
```

```python
plt.plot(x,y,linewidth  = 5)
plt.scatter(points[:,0],points[:,1],alpha=.5)
plt.xlabel('random variable, x',size=14)
plt.ylabel('random variable, y',size=14)
plt.show()
```

```python
selected_points = np.array([points[i] for i in range(len(points)) if points[i,1]<dist(points[i,0])])
```

```python
plt.plot(x,y,linewidth  = 5)
plt.scatter(selected_points[:,0],selected_points[:,1],alpha=.5)
plt.xlabel('random variable, x',size=14)
plt.ylabel('random variable, y',size=14)
plt.show()
```

```python
num_bins = 10
plt.hist(selected_points[:,0],bins=num_bins,density=True, rwidth=0.85,alpha=.5)
plt.plot(x,y,linewidth  = 5,c='C0')
plt.xlabel('random variable, x',size=14)
plt.ylabel('probability distribution',size=14)
plt.show()
```
